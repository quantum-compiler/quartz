```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile, qasm
from qiskit.test.mock import FakeBoeblingen
backend = FakeBoeblingen()
```

```python
barenco_tof_3_circ = QuantumCircuit.from_qasm_file('../circuit/nam-circuits/qasm_files/barenco_tof_3_before.qasm')
barenco_tof_3_circ.draw(output='mpl')
```

```python
optimized_circ = transpile(barenco_tof_3_circ, basis_gates=['rz', 'rx', 'cz'], optimization_level=3)
print(optimized_circ.count_ops())
print(optimized_circ.depth())
optimized_circ.draw(output='mpl')
```
